IMPORTING DEPENDENCIES

```python
import numpy as np
import pandas as pd
import matplotlib as plt
import seaborn as sns
```

Loading Datasets

```python
#importing train dataset

train_data = pd.read_csv("/train (1).csv")
train_data.head(3)
```

```python
#importing test dataset

test_data = pd.read_csv("/test.csv")
test_data.head(3)
```

% Men and Women survived

```python
#calculating percent women survived on titanic


women_survived = train_data[(train_data['Sex'] == 'female') & (train_data['Survived'] == 1)]
rate_women = len(women_survived)/len(train_data[train_data['Sex'] == 'female']) *100
print('% of women survived :',rate_women)
```

```python
#calculating percent men survived on titanic


men_survived = train_data[(train_data['Sex'] == 'male') & (train_data['Survived'] == 1)]
rate_men = len(men_survived)/len(train_data[train_data['Sex'] == 'male']) *100
print('% of men survived :',rate_men)
```

MODEL TRAINING

```python
#importing random forest classifier
from sklearn.ensemble import RandomForestClassifier

#target variable
y = train_data['Survived']

#features
features = ['Pclass','Sex','SibSp','Parch']
X = pd.get_dummies(train_data[features])
X_test = pd.get_dummies(test_data[features])

#model

model = RandomForestClassifier(n_estimators = 100,max_depth = 5,random_state = 1)

#fitting the model
model.fit(X,y)

#predicting the model
predictions = model.predict(X_test)

#making an output dictionary to dataframe
output = pd.DataFrame({'PassengerId': test_data.PassengerId, 'Survived': predictions})

#saving output to a csv file without index with a message
output.to_csv('Submissions.csv',index = False)
print('Your submission was successfully saved!')
```
